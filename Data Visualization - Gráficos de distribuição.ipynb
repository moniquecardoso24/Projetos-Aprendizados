### Passo 3 - Investigando as diferenças

Criar dois histogramas que mostram a distribuição em valores para 'Área (média)', separadamente para tumores benignos e malignos. (Para facilitar a comparação, crie uma única figura contendo os dois histogramas)

```python
# Histogramas - sns.histplot

sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')
```

**Pergunta:**
O tumores malignos têm valores maiores ou menores para 'Área (média)' (relativo a tumores benignos), em média?
O histograma para tumores malignos aparece principalmente à esquerda ou à direita do histograma para tumores benignos? Qual histograma aparece mais largo?

**Resposta:**
Os tumores malignos têm valores mais altos para 'Área (média)', em média. Os tumores malignos têm uma gama maior de valores potenciais. Pode-se observar valores para Area(mean) de acima de 1000 a 2500(eixo x). O histograma para tumores malignoos aparece à direita do benigno, sendo que o maligno é o histograma mais largo no eixo x

### Passo 4 - Verificando a coluna  Radius (worst)

Use os dados do arquivo para criar dois gráficos do KDE que mostram a distribuição em valores para `'Radius (worst)'`, separadamente para tumores benignos e malignos. (Para facilitar a comparação, crie uma única figura contendo os dois gráficos do KDE na mesma célula de código)


```python
# KDE plots - para tumores malignos e benignos - sns.kdeplot

sns.kdeplot(data=cancer_data, x='Radius (worst)',hue='Diagnosis', shade=True)

#shade - preencher as linhas - área pintada
```

**Pergunta:** Um hospital começou recentemente a usar um algoritmo que pode diagnosticar tumores com alta precisão. Dado um tumor com um valor de `'Radius(worst)'` de 25, você acha que o algoritmo tem maior probabilidade de classificar o tumor como benigno ou maligno?

**Resposta:** Conforme mostra o gráfico, o algoritmo tem maior probabilidade de classificar o tumor como maligno. Isso ocorre porque a curva para tumores malignos é muito maior do que a curva para tumores benignos em torno de um valor de 25, no eixo x.
Também, um algoritmo que obtém alta precisão provavelmente tomará decisões com base nesse padrão nos dados.
